```python
import xlrd

data = xlrd.open_workbook(r'C:\Users\zm\Desktop\第一批复工\磷铵事业一公司疫情花名册.xlsx')
table = data.sheets()[0]

tables = []
def import_excel(excel):
    for rown in range(excel.nrows):
        array = {'姓名':'','车间':'','宿舍':'','手机号':'','身份证号':'','家庭住址':''}
        array['姓名']=table.cell_value(rown,0)
        array['车间']=table.cell_value(rown,1)
        array['宿舍']=table.cell_value(rown,3)
        array['手机号']=table.cell_value(rown,4)
        array['身份证号']=table.cell_value(rown,5)
        array['家庭地址']=table.cell_value(rown,6)
        tables.append(array)

print(table.nrows)
print(table.ncols)
print(table.col_values(4))
                 
if __name__ == '__main__':
    import_excel(table)
    for i in tables:
        print(i)
        
                 
                 


   
```
