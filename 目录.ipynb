```python
import os
result = []
def get_all(cwd):
    get_dir = os.listdir(cwd)
    for i in get_dir:
        sub_dir = os.path.join(cwd,i)
        if os.path.isdir(sub_dir):
            get_all(sub_dir)
        else:
            result.append(i)

if __name__ == "__main__":
    get_all(r'F:\环境检查')
    print(result)
   
```
